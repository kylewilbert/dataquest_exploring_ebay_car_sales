```python
# clean `price`column
autos['price'] = (autos['price']
                  .str.replace("$","")
                  .str.replace(',','')
                  .astype(int)
                 )
```

```python
# clean `odometer` column
autos['odometer'] = (autos['odometer']
                    .str.replace('km','')
                    .str.replace(',','')
                    .astype(int)
                    )
```

```python
# rename `odometer` column to `odometer_km`
autos.rename(columns={'odometer':'odometer_km'},inplace=True)
```

## Exploring Price by Brand

```python
autos['brand'].value_counts(normalize=True).sort_values(ascending=False)
```

- looks like it'd be good to take a peek at the brands that have 5% or more percent of the listings

```python
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index
print(common_brands)                    
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_price = brand_only['price'].mean()
    brand_mean_prices[brand] = int(mean_price)
    
brand_mean_prices
```

- Audi, Mercedes-Benz, and BMW are in a class of their own (more than 1.5x the price of Volkswagen)
- Ford and Opel are at the bottom of the top 5%
- Volkswagen is right in the middle

## Exploring mileage

```python
brand_mean_mileage = {}

for brand in common_brands:
    brand_only = autos[autos['brand']== brand]
    mean_mileage = brand_only['odometer_km'].mean()
    brand_mean_mileage[brand] = int(mean_mileage)
    
mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_prices = pd.Series(brand_mean_prices).sort_values(ascending=False)
```

```python
brand_info = pd.DataFrame(mean_mileage, columns=['mean_mileage'])
```

```python
brand_info['mean_price'] = mean_prices
```

```python
brand_info
```

- the mean mileages do not vary as much as the mean prices between brands
